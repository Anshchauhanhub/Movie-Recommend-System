```python
import ast

def convert(obj):
    try:
        L = []
        if obj and obj != '[]': 
            for i in ast.literal_eval(obj):
                L.append(i['name'])
        return L
    except (ValueError, SyntaxError):
        return []


movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
            if counter != 3:
                L.append(i['name'])
                counter += 1
            else:
                break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
import ast

def fetch_director(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
            if i['job'] == "Director":
                L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
import pandas as pd
import numpy as np

# Your existing DataFrame
# Example:
# movies = pd.DataFrame({
#    'overview': ['This is a test.', 'Another test.', np.nan, 'More testing.']
# })

movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else x)
print(movies)
```

```python
# Corrected code for each column
movies['genres'] = movies['genres'].apply(lambda x: [i.replace('"', '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace('"', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace('"', '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace('"', '') for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['genres']+movies['keywords']+movies['cast']+movies['crew'] + movies['overview']
```

```python
from nltk.stem.porter import PorterStemmer
import pandas as pd


ps = PorterStemmer()


def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)


df['tags'] = df['tags'].apply(stem)


print(df['tags'])
```

```python



df.loc[:, 'tags'] = df['tags'].apply(lambda x: x.lower() if isinstance(x, str) else x)

# Print the 'tags' column to verify
print(df['tags'][0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np


df['tags'].fillna('', inplace=True)

cv = CountVectorizer(max_features=5000, stop_words='english')

vectorized_data = cv.fit_transform(df['tags'])


vectorized_array = vectorized_data.toarray()
print(vectorized_array.shape)
```

```python
vectorized_array
```

```python
vectorized_array[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity


similarity = cosine_similarity(vectorized_array)


similarity[0]
```

```python
sorted_similarities = sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
print(sorted_similarities)
```

```python
def recommend(movie):

    movie_index = df[df['title'] == movie].index[0]
    
  
    distances = similarity[movie_index]
    
    
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
   
    for i in movies_list:
        print(df.iloc[i[0]]['title'])

```

```python
recommend('Harry Potter and the Half-Blood Prince')
```
